```python
# content 컬럼 옆에 content_noun 컬럼을 새로 만드는 코드
import pandas as pd
from konlpy.tag import Mecab

# 데이터 불러오기
df = pd.read_csv('/content/전남소멸_10000.csv')

# 형태소 분석기 준비
try:
    mecab = Mecab()
    def extract_nouns(text):
        return ' '.join([word for word, pos in mecab.pos(str(text)) if pos in ['NNG', 'NNP', 'NNB', 'SL', 'SN']])
except:
    from konlpy.tag import Okt
    okt = Okt()
    def extract_nouns(text):
        return ' '.join(okt.nouns(str(text)))

# content_noun 컬럼 생성 (content는 그대로 둠)
df['content_noun'] = df['content'].apply(extract_nouns)

# 결과 저장
df.to_csv('/content/전남소멸_10000_noun.csv', index=False)
```
