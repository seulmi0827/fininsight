```python
import pandas as pd
import re

# 데이터 로드
file_path = '/content/dictionary_sample_dataset(환경).xlsx'
data = pd.read_excel(file_path)

# 핵심 단어 추출 함수 정의
def extract_aspect_terms(text):
    # '을/를', '이/가', '에' 등 조사 앞의 단어를 추출
    matches = re.findall(r'([가-힣]+?)\s*(?:을|를|이|가|에|의)', text)
    # 필터링: 너무 짧거나 일반적인 단어 제외
    core_terms = [word for word in matches if len(word) > 1 and word not in ['환경', '문제', '사업']]
    return ", ".join(core_terms)  # 쉼표로 연결된 핵심 단어 반환

# 각 행의 content 열에서 핵심 단어 추출
data['aspect_term'] = data['content'].apply(extract_aspect_terms)

# 결과 확인
print(data.head())

# 결과를 새로운 엑셀 파일로 저장
output_file_path = 'updated_dictionary_sample_dataset.xlsx'
data.to_excel(output_file_path, index=False)
print(f"추출된 aspect_term이 저장된 파일: {output_file_path}")
```

```python
def extract_aspect_terms(text):
    # 형태소 분석 결과에서 명사(NNG, NNP)와 동사(VV) 추출
    tokens = mecab.pos(text)
    aspect_terms = [word for word, pos in tokens if pos in ['NNG', 'NNP', 'VV']]
    return ", ".join(aspect_terms)  # 쉼표로 연결된 핵심 단어 반환

# 데이터셋에 Aspect Term 추가
data['aspect_term'] = data['content'].apply(extract_aspect_terms)
```

```python
sentiment_mapping = {'긍정': 2, '중립': 1, '부정': 0}
data['sentiment'] = data['sentiment'].map(sentiment_mapping)
```

```python
def prepare_input(text, aspect_term):
    # 형태소 분석된 문장 생성
    analyzed_text = " ".join([word for word, pos in mecab.pos(text)])
    return f"{analyzed_text} [SEP] {aspect_term}"
```

```python
# NaN 값 개수 확인
print("결측값 개수:", data['content'].isnull().sum())

# 결측값이 있는 행 샘플 출력
print("\n결측값 샘플:")
print(data[data['content'].isnull()].head())
```

```python
# 결측값을 빈 문자열로 대체
data['content'] = data['content'].fillna('')

# 또는 결측값이 있는 행 제거
# data = data.dropna(subset=['content'])
```

```python
from konlpy.tag import Mecab

mecab = Mecab()

def generate_aspect_terms(text):
    # 문장에서 명사 추출 (NNG: 일반명사, NNP: 고유명사)
    nouns = mecab.nouns(text)
    return ", ".join(nouns[:3])  # 상위 3개 명사 선택

data['aspect_term'] = data['content'].apply(generate_aspect_terms)
```

```python
# Aspect Term이 없는 행 제외
filtered_data = data[data['aspect_term'].notnull() & (data['aspect_term'] != '')].copy()
print("필터링된 데이터 수:", len(filtered_data))
```

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch

# 모델 및 토크나이저 로드
model = BertForSequenceClassification.from_pretrained('klue/bert-base', num_labels=3)
model.load_state_dict(torch.load('/content/model_state_dict.pth'))
tokenizer = BertTokenizer.from_pretrained('klue/bert-base')

def analyze_sentiment(text, aspect_term):
    # 입력 형식: "[CLS] 문장 [SEP] aspect_term [SEP]"
    combined = f"{text} [SEP] {aspect_term}"

    # 토큰화
    inputs = tokenizer(
        combined,
        padding='max_length',
        truncation=True,
        max_length=128,
        return_tensors='pt'
    )

    # 추론
    with torch.no_grad():
        outputs = model(**inputs)

    # 결과 매핑
    return {0: '부정', 1: '중립', 2: '긍정'}[torch.argmax(outputs.logits).item()]

# 감성 분석 실행
filtered_data['sentiment'] = filtered_data.apply(
    lambda x: analyze_sentiment(x['content'], x['aspect_term']),
    axis=1
)
```
